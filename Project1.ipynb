<b>Time Basis study of the gender by Sport: Athletics</b>

```python
query = '''
SELECT a.Year, a.Season, a.City, a.Sport,
count(Distinct a.Name) as Total_Players, aux.Total_Players_M,auxb.Total_Players_F
FROM athlete_events a
LEFT JOIN
    (
    SELECT Year, Season, City, Sport, Sex, count(Distinct Name) as Total_Players_M
    FROM athlete_events
    Where Sex like 'M'
    and season like ('%Summer%') 
    and Sport like ('%Athletics%')
    Group by Year, Season, City, Sport, Sex
    ) aux
on a.Year = aux.year
    and a.Season = aux.Season
    and a.City = aux.City
    and a.Sport = aux.Sport
LEFT JOIN
    (
    SELECT Year, Season, City, Sport, Sex, count(Distinct Name) as Total_Players_F
    FROM athlete_events
    Where Sex like 'F'
    and season like ('%Summer%') 
    and Sport like ('%Athletics%')
    Group by Year, Season, City, Sport, Sex
) auxb
on a.Year = auxb.year
    and a.Season = auxb.Season
    and a.City = auxb.City
    and a.Sport = auxb.Sport
Where a.season like ('%Summer%') and a.Sport like ('%Athletics%')
Group by a.Year, a.Season, a.City, a.Sport, aux.Total_Players_M, auxb.Total_Players_F
Order by a.Year;
'''
mysql(query)
```

```python
df=mysql(query)
df.plot(x ='Year', y=['Total_Players','Total_Players_M','Total_Players_F'], kind = 'bar')
plt.show()
```

```python
query3 = '''
SELECT a.NOC,b.region, a.Season, a.Sport, a.Sex, count(Distinct a.Name) as Total_Players_F,
round(cast(count(Distinct Name) AS FLOAT)/
(SELECT count(Distinct Name) as Sum_Players_F
FROM athlete_events
Where Sex like 'F'
    and season like ('%Summer%') 
    and Sport like ('Gymnastics%'))*100 ,2)as 'Ratio %'  
FROM athlete_events a
inner join noc_regions b
    on a.NOC = b.NOC
Where a.Sex like 'F'
    and a.season like ('%Summer%') 
    and a.Sport like ('Gymnastics%')
Group by a.NOC,b.region, a.Season, a.Sport, a.Sex
Order By Total_Players_F desc
Limit 10;
'''
mysql(query3)
```

```python
query4 = '''
    
SELECT a.NOC,b.region, a.Season, a.Sport, a.Sex, count(Distinct a.Name) as Total_Players_F,
Round(cast(count(Distinct Name) AS FLOAT)/
(SELECT count(Distinct Name) as Sum_Players_F
FROM athlete_events
Where Sex like 'F'
    and season like ('%Summer%') 
    and Sport like ('%Athletics%'))*100 ,2 )as 'Ratio %'  
FROM athlete_events a
inner join noc_regions b
    on a.NOC = b.NOC
Where a.Sex like 'F'
    and a.season like ('%Summer%') 
    and a.Sport like ('%Athletics%')
Group by a.NOC,b.region, a.Season, a.Sport, a.Sex
Order By Total_Players_F desc
Limit 10;
'''
mysql(query4)
```

```python
df3=mysql(query3)
df4=mysql(query4)
```

```python
plt.pie(df3['Ratio %'], labels = df3['Ratio %'])
plt.legend(df3['region'], bbox_to_anchor=(1.1,1))
plt.title("Global Woman % for Gymnastics, top 10 countries", fontname='DejaVu Sans', 
          fontsize=14,fontstyle='italic',fontweight='bold')
plt.show() 
```
